#### anonymize

```python
fake_name_counter = count(1)

def generate_fake_name(original_name):
    idx = next(fake_name_counter)
    return f"Patient_{idx:03d}"
```

```python
def redact_names(text, name_list):
    if not isinstance(text, str):
        return text
    for name in name_list:
        escaped = re.escape(name.strip())
        text = re.sub(escaped, "[NAME]", text, flags=re.IGNORECASE)
    return text
```

```python
df['anon_name'] = df['name'].apply(generate_fake_name)
df.drop(columns=['name'], inplace=True)
```

```python
name_list = df['anon_name'].unique().tolist()
for col in ['gross_description', 'microscopic_description', 'full_description']:
    df[col] = df[col].apply(lambda x: redact_names(x, name_list))
```
